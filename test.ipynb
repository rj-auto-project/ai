```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pyautogui
import easyocr
import cv2
import re
from dotenv import load_dotenv
import os

load_dotenv()

# Coordinates for screenshot
x1, y1 = 1345, 610  # Top-left
x2, y2 = 1668, 672  # Bottom-right

# Webpage details
url = "https://sso.rajasthan.gov.in/signin?encq=m0ZUFHLqc4t+0vQu27K7jl5cOBbodS7JFafFdflRFZs="
username = os.getenv("username")
password = os.getenv("password")
username_field_id = "cpBodyMain_cpBody_txt_Data1"
password_field_id = "cpBodyMain_cpBody_txt_Data2"
captcha_field_id = "cpBodyMain_cpBody_ssoCaptcha_txtCaptcha"
submit_button_id = "cpBodyMain_cpBody_btn_Login"
driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)
time.sleep(2)

def capture_screenshot(x1, y1, x2, y2, save_path='/home/annone/ai/rajkaj/data/otp.png'):
    width = x2 - x1
    height = y2 - y1
    screenshot = pyautogui.screenshot(region=(x1, y1, width, height))
    screenshot.save(save_path)
    print(f"Screenshot saved as {save_path}")

def read_captcha():
    img = cv2.imread("/home/annone/ai/rajkaj/data/otp.png")
    img = cv2.resize(img, (323, 93))
    cv2.imwrite("/home/annone/ai/rajkaj/data/ww.png",img)
    reader = easyocr.Reader(['en'])
    result = reader.readtext(img, threshold=0.2)
    otp = ''
    for line in result:
        otp += line[1]
    # otp = re.sub(r"[^\w\s]", "", otp)
    print(otp)
    return otp

def login():
    print(username)
    print(password)
    # Input the username
    username_field = driver.find_element(By.ID, username_field_id)
    username_field.send_keys(username)

    # Input the password
    password_field = driver.find_element(By.ID, password_field_id)
    password_field.send_keys(password)

    # Capture screenshot of the captcha
    capture_screenshot(x1, y1, x2, y2, save_path='/home/annone/ai/rajkaj/data/otp.png')

    # Read the captcha from the screenshot
    captcha = read_captcha()
    while True:
        if captcha != '':
            break
        else:
            captcha = read_captcha()
    # Input the captcha
    captcha_field = driver.find_element(By.ID, captcha_field_id)
    captcha_field.send_keys(captcha)

    # Submit the form
    submit_button = driver.find_element(By.ID, submit_button_id)
    submit_button.click()

def navigate_to_table():
    e_file_btn = driver.find_element(By.ID, "zk_comp_146-cave")
    e_file_btn.click()
    file_list_parent_class = driver.find_element(By.ID, "zk_comp_558-body")
    file_class = file_list_parent_class.find_element(By.CLASS_NAME,"z-row")
    file_class.click()
    send_btn = driver.find_element(By.ID, "zk_comp_409-b")
    send_btn.click()

def send_file():
    target_ids = os.getenv("reciever_id")
    navigate_to_table()
    i = 0
    while True:
        target_id = target_ids[i]
        table_parent_element = driver.find_element(By.ID, "zk_comp_557")
        rows = table_parent_element.find_elements(By.CLASS_NAME, 'z-row')
        for row in rows:
            id_element = row.find_element(By.XPATH, ".//td[2]/div/span")
            row_id = id_element.text
            if row_id == target_id:
                radio_button = row.find_element(By.XPATH, ".//td[1]/div/span/input[@type='radio']")
                radio_button.click()
                print(f"Sent to ID : {row_id}")
                send_btn = driver.find_element(By.ID, "zk_comp_1031")
                send_btn.click()
        else:
            print(f"ID {target_id} not found.")
        navigate_to_table()
        if i == 1:
            i = 0
        else:
            i = 1

send_file()
print("sucess")
time.sleep(2)
```

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get("https://sso.rajasthan.gov.in/signin?encq=m0ZUFHLqc4t+0vQu27K7jl5cOBbodS7JFafFdflRFZs=")

# The word you want to search and click
search_word = "Login"
# time.sleep(5)
try:
    # Locate the element containing the specific text and click it
    element = driver.find_element("xpath", f"//*[contains(text(), '{search_word}')]")
    element.click()
    print(f"Clicked on the element containing the word '{search_word}'!")
except NoSuchElementException:
    print(f"Word '{search_word}' not found on the page.")
except ElementClickInterceptedException:
    print("Could not click the element - it might be obscured by another element.")

# Close the driver
# driver.quit()
```
